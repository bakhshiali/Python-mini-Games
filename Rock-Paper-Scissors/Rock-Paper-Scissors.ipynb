```python
'''
function base
'''
import tkinter as tk
import random

# Create the main window
root = tk.Tk()
root.title("Rock-Paper-Scissors Game")

# Create the label for displaying the result
result_label = tk.Label(root, text="Choose your move!", font=("Arial", 18))
result_label.pack(pady=20)

# Create the buttons for the game
rock_button = tk.Button(root, text="Rock", font=("Arial", 14), width=10, height=2)
paper_button = tk.Button(root, text="Paper", font=("Arial", 14), width=10, height=2)
scissors_button = tk.Button(root, text="Scissors", font=("Arial", 14), width=10, height=2)

# Function to determine the winner of the game
def determine_winner(player_move, computer_move):
    if player_move == computer_move:
        return "Tie game!"
    elif (player_move == "Rock" and computer_move == "Scissors") or (player_move == "Paper" and computer_move == "Rock") or (player_move == "Scissors" and computer_move == "Paper"):
        return "You win!"
    else:
        return "You lose!"

# Function to handle button clicks
def button_click(player_move):
    # Generate a random move for the computer
    moves = ["Rock", "Paper", "Scissors"]
    computer_move = random.choice(moves)

    # Determine the winner of the game
    result = determine_winner(player_move, computer_move)

    # Update the label with the result
    result_label.config(text=f"You chose {player_move}.\nComputer chose {computer_move}.\n{result}")

# Add the button click handlers
rock_button.config(command=lambda: button_click("Rock"))
paper_button.config(command=lambda: button_click("Paper"))
scissors_button.config(command=lambda: button_click("Scissors"))

# Pack the buttons
rock_button.pack(side=tk.LEFT, padx=20)
paper_button.pack(side=tk.LEFT, padx=20)
scissors_button.pack(side=tk.LEFT, padx=20)

# Start the main loop
root.mainloop()
```

```python
'''
class base
'''
import tkinter as tk
import random

class RockPaperScissors:
    def __init__(self, master):
        self.master = master
        master.title("Rock-Paper-Scissors Game")
        master.geometry("500x300")
        master.resizable(width=False, height=False)

        # Create the label for displaying the result
        self.result_label = tk.Label(master, text="Choose your move!", font=("Arial", 18))
        self.result_label.pack(pady=20)

        # Create the buttons for the game
        self.rock_button = tk.Button(master, text="Rock", font=("Arial", 14), width=10, height=2, command=lambda: self.button_click("Rock"))
        self.paper_button = tk.Button(master, text="Paper", font=("Arial", 14), width=10, height=2, command=lambda: self.button_click("Paper"))
        self.scissors_button = tk.Button(master, text="Scissors", font=("Arial", 14), width=10, height=2, command=lambda: self.button_click("Scissors"))

        # Pack the buttons
        self.rock_button.pack(side=tk.LEFT, padx=20)
        self.paper_button.pack(side=tk.LEFT, padx=20)
        self.scissors_button.pack(side=tk.LEFT, padx=20)

    def determine_winner(self, player_move, computer_move):
        if player_move == computer_move:
            return "Tie game!"
        elif (player_move == "Rock" and computer_move == "Scissors") or (player_move == "Paper" and computer_move == "Rock") or (player_move == "Scissors" and computer_move == "Paper"):
            return "You win!"
        else:
            return "You lose!"

    def button_click(self, player_move):
        # Generate a random move for the computer
        moves = ["Rock", "Paper", "Scissors"]
        computer_move = random.choice(moves)

        # Determine the winner of the game
        result = self.determine_winner(player_move, computer_move)

        # Update the label with the result
        self.result_label.config(text=f"You chose {player_move}.\nComputer chose {computer_move}.\n{result}")

root = tk.Tk()
app = RockPaperScissors(root)
root.mainloop()
```
