```python
import tkinter as tk
import random

# Initial parameters
board_Width = 400
board_Height = 400
snake_Speed = 200
SPACE_SIZE = 20
BODY_SIZE = 2
SNAKE = "#0000F3"
FOOD = "#FFFFFF"
BACKGROUND = "#000000"
score = 0
direction = 'Down'

# Create the window
root = tk.Tk()
root.title("Snake")

# Create the canvas
canvas = tk.Canvas(root, bg="black",height=board_Height, width=board_Width)
canvas.pack()
canvas.create_text(board_Width/2, 100, text="Scores:{}".format(score),
                       font=("consolas", 20), fill="white", tag="score")

# Class to design the snake
class Snake:
    def __init__(self):
        self.body_size = BODY_SIZE
        self.coordinates = []
        self.squares = []
        for i in range(0, BODY_SIZE):
            self.coordinates.append([SPACE_SIZE, SPACE_SIZE])

        for x, y in self.coordinates:
            square = canvas.create_rectangle(x, y, x + SPACE_SIZE, y + SPACE_SIZE,fill=SNAKE, tag="snake")
            self.squares.append(square)

# Class to design the food
class Food:
    def __init__(self):
        x = random.randint(0,(board_Width / SPACE_SIZE)-1) * SPACE_SIZE
        y = random.randint(0,(board_Height / SPACE_SIZE) - 1) * SPACE_SIZE
        self.coordinates = [x, y]
        canvas.create_oval(x, y, x + SPACE_SIZE, y +SPACE_SIZE, fill=FOOD, tag="dot")
        
# Function to check the next move of snake
def next_turn(snake, food):
    x, y = snake.coordinates[0]
    if direction == "Up":
        y -= SPACE_SIZE
    elif direction == "Down":
        y += SPACE_SIZE
    elif direction == "Left":
        x -= SPACE_SIZE
    elif direction == "Right":
        x += SPACE_SIZE
    snake.coordinates.insert(0, (x, y))
    square = canvas.create_rectangle(x, y, x + SPACE_SIZE,y + SPACE_SIZE, fill=SNAKE)
    snake.squares.insert(0, square)
    
    if x == food.coordinates[0] and y == food.coordinates[1]:
        global score
        score += 1
        #label.config(text="Points:{}".format(score))
        canvas.itemconfigure('score', text="Scores:{}".format(score))
        canvas.delete("dot")
        food = Food()
    else:
        del snake.coordinates[-1]
        canvas.delete(snake.squares[-1])
        del snake.squares[-1]
    if check_collisions(snake):
        game_over()
    else:
        root.after(snake_Speed, next_turn, snake, food)

# Function to control direction of snake
def change_direction(event):#new_direction):
    global direction
    if event.keysym == 'Left':
        if direction != 'Right':
            direction = event.keysym
    elif event.keysym == 'Right':
        if direction != 'Left':
            direction = event.keysym
    elif event.keysym == 'Up':
        if direction != 'Down':
            direction = event.keysym
    elif event.keysym == 'Down':
        if direction != 'Up':
            direction = event.keysym
    elif event.keysym == 'space':
        direction ='Down'

# function to check snake's collision and position
def check_collisions(snake):
    x, y = snake.coordinates[0]
    # out of borders
    if x < 0 or x >= board_Width:
        return True
    elif y < 0 or y >= board_Height:
        return True
    # self-collision
    for body_part in snake.coordinates[1:]:
        if x == body_part[0] and y == body_part[1]:
            return True
    return False
def start_game():
    snake = Snake()
    food = Food()
    #next_turn(snake, food)
    #restart the game
    root.after(snake_Speed, next_turn, snake, food)

# Function to control everything
def game_over():
    canvas.delete('all')
    
    canvas.create_text(canvas.winfo_width()/2,canvas.winfo_height()/3,font=('consolas', 30),text="GAME OVER", fill="red",tag="gameover")
    canvas.create_text(canvas.winfo_width()/2,canvas.winfo_height()/2,font=('consolas', 20),text="Press SPACE to start over.", fill="red",tag="gameover")

# Giving title to the gaming window
def reset_game(event):
    global score
    if event.keysym == 'space':
        score=0
        canvas.delete('all')
        #recreate the score text
        canvas.create_text(board_Width/2, 100, text="Scores:{}".format(score),
                           font=("consolas", 20), fill="white", tag="score")
        start_game()


start_game()
# Display of Points Scored in Game
root.update()
window_width = root.winfo_width()
window_height = root.winfo_height()
x = int((root.winfo_screenwidth()/2) - (window_width/2))
y = int((root.winfo_screenheight()/2) - (window_height/2))
root.geometry(f"{window_width}x{window_height}+{x}+{y}")

canvas.focus_set()

#key bind to functions
root.bind("<Key>", change_direction)
# Bind the reset_game function to the space bar
root.bind("<Key>", reset_game,add="+")#"<space>", lambda event: reset_game())

root.mainloop()
```
